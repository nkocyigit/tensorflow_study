```python
import sklearn as sklearn
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
# for plotting on the current document
%matplotlib inline
```

```python
# number of data points
DATA_SIZE = 80
# generate uniform noise in given boundaries
noise = np.random.uniform(low=-25, high=25, size=(DATA_SIZE,))
# X is a sequential increasing data
X = np.arange(DATA_SIZE)
# define Y and YNoisy
Y = np.arange(DATA_SIZE)
YNoisy = np.arange(DATA_SIZE)

# in a loop add noise to the data
# main model is y = 5 + 3x 
for i in range(DATA_SIZE):
    YNoisy[i]  = 5 + 3 * X[i] + noise[i]
    Y[i] = 5 + 3 * X[i]

# convert to 1 column and n rows shape
X = X.reshape(-1,1)
YNoisy = YNoisy.reshape(-1,1)

# we will use linear regression
linReg = LinearRegression()
linReg.fit(X, YNoisy)

# print results
print(linReg.intercept_)
print(linReg.coef_)

# predicted values after fit
YPred = np.arange(DATA_SIZE)
for i in range(DATA_SIZE):
    YPred[i] = linReg.coef_ * X[i] + linReg.intercept_

# mean squared error
mse = sklearn.metrics.mean_squared_error(YNoisy,YPred)
print("MSE : ",mse)
    
# plot all data
plt.plot(X,YNoisy,"b.",X,noise,X,YPred)

# print mean for noise data
print(np.mean(noise))
# print std for noise data
print(np.std(noise))
```
