```python
# linear regression using sklearn
import sklearn as sklearn
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
# for plotting on the current document
%matplotlib inline
```

```python
# number of data points
DATA_SIZE = 80
# generate uniform noise in given boundaries
noise = np.random.uniform(low=-25, high=25, size=(DATA_SIZE,))
# X is a sequential increasing data
X = np.arange(DATA_SIZE)
# define Y and YNoisy
Y = np.arange(DATA_SIZE)
YNoisy = np.arange(DATA_SIZE)

# in a loop add noise to the data
# main model is y = 5 + 3x 
for i in range(DATA_SIZE):
    YNoisy[i]  = 5 + 3 * X[i] + noise[i]
    Y[i] = 5 + 3 * X[i]

# convert to 1 column and n rows shape
X = X.reshape(-1,1)
YNoisy = YNoisy.reshape(-1,1)

# we will use linear regression
linReg = LinearRegression()
linReg.fit(X, YNoisy)

# print results
print(linReg.intercept_)
print(linReg.coef_)

# predicted values after fit
YPred = np.arange(DATA_SIZE)
for i in range(DATA_SIZE):
    YPred[i] = linReg.coef_ * X[i] + linReg.intercept_

# mean squared error
mse = sklearn.metrics.mean_squared_error(YNoisy,YPred)
print("MSE : ",mse)
    
# plot all data
plt.plot(X,YNoisy,"b.",X,noise,X,YPred)

# print mean for noise data
print("Noise Mean : ",np.mean(noise))
# print std for noise data
print("Noise Std : ",np.std(noise))
```

```python
# linear regression using the tensorflow
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
# for plotting on the current document
%matplotlib inline
```

```python
# define the model parameters
# model is Y = 5 + 3X
# we will add some uniform noise to the model
# Y = theta0 + theta1 * X + noise
# theta0 and theta1 are model parameters (variables)
# X and Y are placeholders (model input-output)

# number of data points
DATA_SIZE = 80
# print cost and thetas every epoch size iteration
EPOCH_SIZE = 50
# gradient descent algorithm learning rate
LEARNING_RATE = 0.0001
# generate uniform noise in given boundaries
noise = np.random.uniform(low=-25, high=25, size=(DATA_SIZE,))
# x train is a sequential increasing data
x_train = np.arange(DATA_SIZE)
# define y train
y_train = np.arange(DATA_SIZE)

# in a loop add noise to the data
# main model is Y = 5 + 3X
# generate y train values adding noise
for i in range(DATA_SIZE):
    y_train[i] = 5 + 3 * x_train[i] + noise[i]
    
# theta0 and theta1 are model parameters (variables)
theta0 = tf.Variable([.0], dtype=tf.float32)
theta1 = tf.Variable([.0], dtype=tf.float32)

# X and Y are placeholders (model input-output)
X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)

linear_model = theta0 + theta1 * X

# loss (sum of squares)
loss = tf.reduce_sum(tf.pow(linear_model-Y, 2)) / (2 * DATA_SIZE)

# optimizer function
optimizer = tf.train.GradientDescentOptimizer(LEARNING_RATE)
train = optimizer.minimize(loss)

# init all variables
init = tf.global_variables_initializer()

# create the session
sess = tf.Session()
# initialize the variables
sess.run(init)
# run training
for i in range(10000):
    sess.run(train, {X: x_train, Y: y_train})
    if i % EPOCH_SIZE == 0 :
        print("Epoch:", '%04d' % (i))
        curr_theta1, curr_theta0, curr_loss = sess.run([theta1, theta0, loss], {X: x_train, Y: y_train})
        print("theta1: %06f theta0: %06f loss: %06f"%(curr_theta1, curr_theta0, curr_loss))

# evaluate training accuracy
curr_theta1, curr_theta0, curr_loss = sess.run([theta1, theta0, loss], {X: x_train, Y: y_train})
print("theta1: %s theta0: %s loss: %s"%(curr_theta1, curr_theta0, curr_loss))

# predicted values after fit
pred = np.arange(DATA_SIZE)
for i in range(DATA_SIZE):
    pred[i] = curr_theta1 * x_train[i] + curr_theta0
    
# plot all data
plt.plot(x_train,y_train,"b.",x_train,noise,x_train,pred)    
```
